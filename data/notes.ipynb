### Thoughts and notes about data
Ideas for meaningful features after studying https://teamepochgithub.github.io/aicup-birds-radar-detection-tool/
- I have skipped descriptions of the features that names are quite clear etc.
- lat, lon fluctuation (max - min) - VERY good at distinguishing Clutter
- PCA explained variance for lat, long, altitudes - My attempt at distinguishing Birds Of Prey which like to make circles. Although I ignore the v1, v2, v3 (look into data.ipnyb) in the end you may find some use for them. There might be a better way to do it than PCA.
- local_circurality_score is similar to above but its look locally at pieces of the trajectory with a sliding window. Reason for this is the loops birds of prey do are usually small at scattered across the whole flight. We then take the max out of the all of the scores. You can consider doing mean instead.
- It actually turns out the PCA is better for disntinguishing Birds of Prey if we ingore the z dimension. Thus the 2d method for this works similarly.
- I have also treated the radar_bird_size as nominal because it was hard to identify 'flock'. You may change this to ordinal if you prefer
- The reason why df_transform() function takes so long is because of A LOT OF PCA fitting. It is not that bad though. Currently, 10s for ~2600 rows.
